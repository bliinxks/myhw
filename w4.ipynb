```python
def print_dict(dict):
  for key in dict:
    if key == 2 or key ==3 or key == 4:
      print(key, ' = ', dict[key], '')
    else:
      print(key, ' = ', dict[key], '')

s=''
dict={}
for i in s:
  if i not in dict:
    dict[i] = 1
  else:
    dict[i] += 1

print_dict(dict)
```

```python
import random
with open('123.txt', 'w') as f:
    N = 20
    mas = [random.randint(-100, 100) for _ in range(N)]
    for num in mas:
        f.write(f"{num}\n")

cnt = 0
for i in range(len(mas)):
  for j in range(i, len(mas)):
      if mas[i] == - mas[j] and mas[i]!= 0:
            cnt+=1
            print(cnt, ' = ', mas[i], ',', mas[j])
```

```python
def find_books_by_author(auth):
    result = []
    with open('bibl.txt', 'r', encoding='utf-8') as f:
        for s in f:
            parts = s.strip().split()
            if len(parts) >= 3:
                author = parts[0]
                title = ' '.join(parts[1:-1])
                year = int(parts[-1])
                if author == auth and year >= 1960:
                    result.append(title)
    return result

author = input(" : ")
books = find_books_by_author(author)
print(f" {author},   1960 :")
for book in books:
    print(f"- {book}")
```

```python
with open('abc.txt', 'r', encoding='utf-8') as f:
    text = f.read()

new_text = []
for sentence in text.split('.'):
    sentence = sentence.strip()
    if not sentence:
        continue

    words = sentence.split()
    unique_words = []
    seen_words = set()

    for word in words:
        clean_word = word.strip('!?,.').lower()
        if clean_word not in seen_words:
            unique_words.append(word)
            seen_words.add(clean_word)

    new_sentence = ' '.join(unique_words)
    new_text.append(new_sentence)

result = '. '.join(new_text) + '.'
with open('new_text.txt', 'w', encoding='utf-8') as f:
    f.write(result)

print("!")
```
