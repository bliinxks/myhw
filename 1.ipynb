```python
import math

def func(x, y):
    cos = y / x
    rad = math.acos(cos)
    deg = math.degrees(rad)
    return rad, deg

k = float(input("enter stick: "))
h = float(input("enter step: "))

y = 2.0
while y <= 3.0:
    rad, deg = func(k, y)
    print(f"{y:.2f}\t{rad:.4f}\t\t{deg:.2f}")
    y += h
```
