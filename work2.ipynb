```python
class CartItem:
  def __init__(self, name, num,price):
      self.name = name
      self.num = num
      self.price = price

  def update_quantity(self, new_quantity):
    if new_quantity > 0:
      self.num = new_quantity
    else:
      print("number can't be <0")

  def total_cost(self):
    return self.num * self.price
```
