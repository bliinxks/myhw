```python
def summ(dict1, dict2):
    sum_dict = {}
    for key in dict1:
        sum_dict[key] = dict1[key]

    for key in dict2:
        if key in sum_dict:
            sum_dict[key] += dict2[key]
        else:
            sum_dict[key] = dict2[key]

    return sum_dict

dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}
print(summ(dict1, dict2))
```

```python
def sort_set(s):
    sorted_list = []
    set1 = s.copy()

    while set1:
        mini = float('inf')
        for num in set1:
            mini = min(mini, num)

        sorted_list.append(mini)
        set1.remove(mini)
    return sorted_list

set1 = {5, 2, 8, 1, 9}
print(sort_set(set1))
```
